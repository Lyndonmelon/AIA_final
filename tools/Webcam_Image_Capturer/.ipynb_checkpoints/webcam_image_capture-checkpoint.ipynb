#### This is a short program that helps capture series of images using webcam/external cam. The program waits 1 second and starts capturing 80 frames (or until user presses 'q'). Currently capturing at 0.08s rate

```python
import cv2
import os
from time import sleep


folder_name = 'test_36'
sav_dir = 'H:\Test\images_' + folder_name # save folder path

if not os.path.exists(sav_dir):
    os.makedirs(sav_dir)

# Use pc webcam to capture 1280x720 images
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)

# Give 1 second to get ready
sleep(1)
count = 0

while count < 80: # capture 80 images in the given time
    # get a frame
    ret, frame = cap.read()
    # show a frame
    cv2.imshow("capture", frame)
    file_name = folder_name + '_' + str(count) + '.jpg' # save file name
    fname = os.path.join(sav_dir, file_name)
    
    cv2.imwrite(fname, frame)
    count += 1
    sleep(0.08) # delay between captures in seconds.
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
